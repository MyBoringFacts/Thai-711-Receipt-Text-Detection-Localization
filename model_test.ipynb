```python
import os
from ultralytics import YOLO

# Load the trained YOLO model
model = YOLO(r"model\best_model.pt")

# Folder containing your images
folder_path = r"New folder"

# Get a list of image files (adjust the file extensions as needed)
image_files = [os.path.join(folder_path, f) for f in os.listdir(folder_path)
               if f.lower().endswith(('.jpg', '.jpeg', '.png'))]


# Process each image
for img_path in image_files:
    print(f"Processing image: {img_path}")
    results = model.predict(source=img_path, imgsz=640, conf=0.25)
    # results is a list; iterate over each result (usually one per image)
    for result in results:
        result.show()  
```
